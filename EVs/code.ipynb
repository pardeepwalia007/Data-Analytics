```python
# first sort the values by year then rename and then convert to dataframe
vehicle_count = (df['Model Year'].value_counts().sort_index()
                 .rename('Vehicle_count')
                 .to_frame())
vehicle_count
```

```python
# YOY absolute basically subtracting prev from current
vehicle_count['YoY_abs'] = vehicle_count['Vehicle_count'].diff()
vehicle_count['YoY_abs']
```

```python
# YoY percentage
prev = vehicle_count['Vehicle_count'].shift(1)
vehicle_count['YoY_pct'] = np.where((prev>0),((vehicle_count['Vehicle_count']-prev)/prev)*100.0,np.nan)
vehicle_count
```

```python
# CAGR Compound Annual Growth Rate , window = hom many years of window eg 2014 ->2025
end_year = vehicle_count.index.max()
window = 11
start_yr = max(year_li, end_year-(window-1))
# window candidate year count and latest year count
c_start = vehicle_count.loc[start_yr,"Vehicle_count"]
c_end = vehicle_count.loc[year_hi,'Vehicle_count']

year_elapsed = year_hi-start_yr

CAGR = np.nan
if year_elapsed>0 and c_start>0 and c_end>0:
    CAGR= (c_end / c_start)**(1.0/year_elapsed)-1.0

CAGR = round(float(CAGR )*100,2)
CAGR
```

```python
df_valid =  df.query("`Model Year`> 2009 and  `Model Year` <2026")
```

```python
# Car companies with null Electric Range
null_ranges=df_valid.query("`Electric Range` == 0.0").loc[:,['Make','Model Year']].value_counts().sort_index().to_frame()
```

```python
null_ranges
```

```python
null_ranges['ttl_cnt '] =df_valid.groupby(['Make','Model Year']).agg(counts = ('VIN (1-10)','count'))
```

```python
null_ranges
```

```python
# ratio of missed Electric Ranges
missing_ratio = round(null_ranges['count']/null_ranges['ttl_cnt '],2)
missing_ratio
```

```python
# These cutoffs are industry-standard heuristics (not arbitrary) — they’re commonly used in credit risk, healthcare, and public datasets when deciding whether to impute vs drop. 0.9, 0.1
null_ranges['missing_ratio'] =missing_ratio
```

```python
cond = [null_ranges['missing_ratio']>=0.9,(null_ranges['missing_ratio']<0.9) &( null_ranges['missing_ratio']>=0.1),null_ranges['missing_ratio']<0.1]
vals=["all_missing","partial_missing","partial_missing"]
null_ranges['imputation_flag'] = np.select(cond,vals,default="none_missing")
```

```python
null_ranges.sample(20)
```

```python
# using the falg from null_ranges data frame

df_valid = df_valid.merge(null_ranges.reset_index()[['Make','Model Year','imputation_flag']],on=['Make','Model Year'], how='left'
)
```

```python
df_valid['imputation_flag'] = df_valid['imputation_flag'].fillna("none_missing")
```

```python
df_valid['imputation_flag'].value_counts()
```

```python
#  df_valid has columns: 'Make','Model','Model Year','Electric Vehicle Type','Electric Range','VIN (1-10)','imputation_flag'
# imputation_flag ∈ {'all_missing','partial_missing','none_missing'}
df_valid['EV_type'] = np.where(df_valid['Electric Vehicle Type'].str.contains('battery',case=False,na=False),"BEV","PHEV")
```

```python
# data lineage / transparency flag
df_valid['zero_range'] = np.where(df_valid['Electric Range'] == 0.0,True,False)

# converts 0.0 to nan 
df_valid['range_clean'] = df_valid['Electric Range'].mask(df_valid['Electric Range']== 0.0).clip(0,600)
```

```python
df_valid['imputation_flag'].value_counts()
```

```python
# Imputation based on the flag type

# Partial missing group
partial_mask  = df_valid['imputation_flag'].eq("partial_missing") & df_valid['zero_range']

# 1) Make + Model + Year + EV_Type

m = partial_mask & df_valid["range_clean"].isna()
df_valid.loc[m, "range_clean"] = (
    df_valid.loc[m]
            .groupby(["Make","Model","Model Year","EV_type"])["range_clean"]
            .transform(lambda s: s.fillna(s.median()))
)
```

```python
# 2) Make + Year + EV_Type
m = partial_mask & df_valid["range_clean"].isna()
df_valid.loc[m, "range_clean"] = (
    df_valid.loc[m]
            .groupby(["Make","Model Year","EV_type"])["range_clean"]
            .transform(lambda s: s.fillna(s.median()))
)
```

```python
# 3) Make + EV_Type (last in-group fallback)
m = partial_mask & df_valid["range_clean"].isna()
df_valid.loc[m, "range_clean"] = (
    df_valid.loc[m]
            .groupby(["Make","EV_type"])["range_clean"]
            .transform(lambda s: s.fillna(s.median()))
)
```

```python
# ALl missing Values Imputation

mask_all = df_valid["imputation_flag"].eq("all_missing") & df_valid["zero_range"]
```

```python
# ALL MISSING: no valid peers → EV-Type global median
# (Still only for rows that were originally missing)
# --------------------------------------------------
type_medians = df_valid.groupby("EV_type")["range_clean"].median()
df_valid.loc[mask_all, "range_clean"] = df_valid.loc[mask_all, "EV_type"].map(type_medians)

# --------------------------------------------------
# TRANSPARENCY FLAGS / QA
# --------------------------------------------------
df_valid["range_imputed"] = df_valid["zero_range"] & df_valid["range_clean"].notna()
df_valid["range_still_na"] = df_valid["range_clean"].isna()  # should be ~0; investigate if >0
```

```python
df_valid['range_still_na'].sum()
```

```python
# Step 1: EV-Type median fallback for ANY NaN
type_medians = df_valid.groupby('EV_type')['range_clean'].median()
mask_any_na = df_valid['range_clean'].isna()
df_valid.loc[mask_any_na, 'range_clean'] = df_valid.loc[mask_any_na, 'EV_type'].map(type_medians)

# Step 2: Global median fallback (for cases with missing EV_Type or EV_Type median also NaN)
overall_med = df_valid['range_clean'].median(skipna=True)
df_valid['range_clean'] = df_valid['range_clean'].fillna(overall_med)

# Step 3: Refresh flags
df_valid['range_still_na'] = df_valid['range_clean'].isna()
df_valid['range_imputed'] = df_valid['zero_range'] & df_valid['range_clean'].notna()
```

```python
df_valid['range_still_na'].sum()
```

```python
df_valid['EV_type'].value_counts()
```

```python
# Long-range flag (>= 300 miles)
df_valid['long_range_flag'] = ((df_valid['range_clean'] >= 300 )& (df_valid['EV_type'].eq('BEV')) ).astype('int8')

# --- Range stats per year ---
kpi4_range_stats = (
    df_valid.groupby(['Model Year','EV_type'])['range_clean']
      .agg(
          count='count',
          median='median',
          p25=lambda s: s.quantile(0.25),
          p75=lambda s: s.quantile(0.75),
          min='min',
          max='max',
          mean='mean'
      )
      .reset_index()
)

# --- Long-range share per year ---
kpi4_long_share = (
    df_valid.groupby(['Model Year','EV_type'])['long_range_flag']
      .mean()
      .mul(100).round(1)
      .reset_index(name='long_range_share_pct')
)

# missing value and imputed value percentage
kpi4_coverage = (
    df_valid.groupby(['Model Year','EV_type'])
      .agg(
          total_vehicles=('VIN (1-10)','count'),
          originally_missing=('zero_range','sum'),
          imputed=('range_imputed','sum')
      )
      .assign(
          originally_missing_pct=lambda x: (x['originally_missing']/x['total_vehicles']*100).round(1),
          imputed_pct=lambda x: (x['imputed']/x['total_vehicles']*100).round(1)
      )
      .reset_index()
)

# --- Final merged KPI #4 table ---
kpi4_final = (
    kpi4_range_stats
      .merge(kpi4_long_share, on=['Model Year','EV_type'], how='left')
      .merge(kpi4_coverage[['Model Year','originally_missing_pct','imputed_pct','EV_type']], on=['Model Year','EV_type'], how='left')
      .sort_values(['Model Year','EV_type'])
)
```

KPI #5



```python
df_valid['Base MSRP'].value_counts()
```

```python
# outlier
df_valid[df_valid['Base MSRP']>800000]
```

```python
# converting 0.0 to numeric nan

df_valid['msrp'] = pd.to_numeric(df_valid['Base MSRP'],errors='coerce')
df_valid['msrp'] = df_valid['msrp'].mask(df_valid['msrp']== 0.0,np.nan)
```

```python
# Bucketting the vehicle based on base msrp
bins = [0,40000,60000,80000,100000,float('inf')]
lables = ['<40k','40–60k','60–80k','80k-90k','100k+']
df_valid['msrp_bucket'] = pd.cut(df_valid['msrp'], bins=bins, labels=lables, include_lowest=True)
df_valid['msrp_bucket'] = df_valid['msrp_bucket'].cat.add_categories(['Unknown'])
df_valid['msrp_bucket'] = df_valid['msrp_bucket'].fillna('Unknown')
```

```python
df_valid['msrp_bucket'].value_counts()
```

```python
# Affordability Distribution
kpi5_affordability= (
    df_valid.groupby(['Model Year','msrp_bucket'])
    .size()
    .reset_index(name='count')
    .pivot_table(index='Model Year',columns='msrp_bucket',values='count')
    .fillna(0)
    .astype(int)
    .sort_index()
)
```

```python
kpi5_affordability
```

```python
# normalizing to pecentages
kpi5_affordability_pct = (
    kpi5_affordability.div(kpi5_affordability.sum(axis=1),axis=0).mul(100).round(1)
)
```

```python
# normalizing county name
df_valid['county_norm'] = (
    df_valid['County'].str.strip().str.title().str.replace(" County","",regex=False)
)
```

```python
kpi6_county = df_valid.groupby(['Model Year','county_norm']).size().reset_index(name='count')
# top counties
top_counties = df_valid['county_norm'].value_counts().head(5).index

kpi6_top_counties = (
    df_valid[df_valid['county_norm'].isin(top_counties)]
    .groupby(['Model Year','county_norm'])
    .size()
    .reset_index(name='count')
    .pivot_table(index = 'Model Year',columns='county_norm',values='count')
    .fillna(0)
    .astype(int)
)
```
